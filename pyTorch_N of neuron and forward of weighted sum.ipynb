#### Input layer to hidden layer 1

```python
#                                        5 neron of h1  from neuron of input layer     
W1 = torch.tensor(np.random.uniform(0,1,(5,inputDim)),requires_grad=True)
```

So we have 5 matrices of nueron in hidden layer 1.

#### Input layer to hidden layer 2

```python
#                                        4 neron of h2 from 5 neuron of h1     
W2 = torch.tensor(np.random.uniform(0,1,(4,5)),requires_grad=True)
```

#### Hidden layer 2 to output layer

```python
W3 = torch.tensor(np.random.uniform(0,1,(2,4)),requires_grad=True)
```

```python
weigth_list = [W1,W2,W3]
```

#### Forward step function

```python
def forwardStep(X,W_list):
    # Input data
    h = torch.from_numpy(X)
    for W in W_list:
        # Weigthed sum function
        h = torch.matmul(W,h)
    return h
```

```python
z = forwardStep(X[0,:],weigth_list)
print(z)
```
