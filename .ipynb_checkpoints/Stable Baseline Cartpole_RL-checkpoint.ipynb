```python
import gym
from stable_baselines3 import PPO # Proximal Policy Optimization (PPO)
# Vectorized Environments are a method for training multiple envionment in vectorized form
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy
```

## Environment

```python
env = gym.make('CartPole-v1')
```

```python
# test environment is working
episodes = 15
for episode in range(1, episodes+1):
    state = env.reset()
    done = False
    score = 0

    while not done:
        env.render()
        action = env.action_space.sample()
        n_state, reward, terminated, truncated, info = env.step(action)
        done = terminated or truncated
        score += reward

    print("Episode: {} Score: {}".format(episode, score))

env.close()
```

```python
env.action_space
```

```python
env.action_space.sample()
```

```python
env.observation_space.sample()
# The result can be discrete or continuous (sometimes called box) value
```

### Training model

```python
env = DummyVecEnv([lambda: env])
# MlpPolicy	Policy object that implements actor critic, using a MLP (2 layers of 64)
model = PPO('MlpPolicy', env, verbose=1) # Check the device is used
```

```python
# Training model with assume timestpe
model.learn(total_timesteps=20000)
```

## Evaluate model

```python
# In google clolab render=False be cuase thecan not render
# render=True
evaluate_policy(model, env, n_eval_episodes=10)
# Ouput is (reward, standard deviation of n of episode)
```

```python
env.close()
```

### Predict model action

```python
episodes = 5
for episode in range(1, episodes+1):
    state = env.reset()
    done = False
    score = 0

    while not done:
        env.render()
        action, _ = model.predict(state)
        state, reward, done, info = env.step(action)
        score += reward

    print("Episode: {} Score: {}".format(episode, score))

env.close()
```

## Train model with Early stoping

```python
# Use callback using reward threshold
from stable_baselines3.common.callbacks import EvalCallback, StopTrainingOnRewardThreshold
```

```python
save_path = r"./"
```

```python
stop_callback = StopTrainingOnRewardThreshold(reward_threshold=195, verbose=1)
eval_callback = EvalCallback(env, callback_on_new_best=stop_callback, eval_freq=10000,
                            best_model_save_path= save_path, verbose=1)
```

```python
model = PPO('MlpPolicy', env, verbose=1)
```

```python
model.learn(total_timesteps=10000, callback=eval_callback)
```

```python
evaluate_policy(model, env, n_eval_episodes=10)
```

```python
env.close()
```

## Changing a policy architecture

```python
# pi = input = [l1,l2,l3,...,ln]
# vf = value funtion = [l1,l2,l3,...,ln]
# In each ln (layer) will store number of neuron in each layer
net_arch = [dict(pi=[64,64,64], vf=[64,64,64])]
```

```python
model = PPO('MlpPolicy', env, verbose=1, policy_kwargs={'net_arch': net_arch})
```

```python
model.learn(total_timesteps=10000, callback=eval_callback)
```

```python
evaluate_policy(model, env, n_eval_episodes=10)
```

```python
env.close()
```
